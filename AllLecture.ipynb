```python
def is_square(x):
    if len(x) == len(x[0]):
        return True
    else:
        return False
    
def is_symmetrical(x):
    if not is_square(x):
        return False
    else:
        for i in range (0,len(x)):
            for j in range (0,len(x[0])):
                if x[i][j] != x[j][i]:
                    return False
        return True
    
def is_diagonal(x):
    if not is_square(x):
        return False
    else:
        for i in range (0,len(x)):
            for j in range (0,len(x[0])):
                if i != j:
                    if x[i][j] != 0:
                        return False
        return True
    
def is_identity(x):
    if not is_square(x):
        return False
    else:
        for i in range (0,len(x)):
            for j in range (0,len(x[0])):
                if i == j:
                    if x[i][j] != 1:
                        return False
                elif i != j:
                    if x[i][j] != 0:
                        return False
        return True

def is_zero(x):
    for i in range (0,len(x)):
        for j in range (0,len(x[0])):
            if x[i][j] != 0:
                return False
    return True
    
def transpose(x):
    tran = []
    for i in range (0,len(x)):
        temp = []
        for j in range (0,len(x[0])):
            temp.append(x[j][i])
        tran.append(temp)
    return tran
```

```python
import numpy as np
def multiply(A,B):
    C = np.zeros(((len(A)),(len(B[0]))),dtype=int)
    i = 0
    if len(A) == len(B[0]):
        while i < len(A):
            j = 0
            while j < len(B[0]):
                sum = 0
                k = 0
                while k < len(A):
                    mul = A[i][k] * B[k][j]
                    sum = sum + mul
                    k += 1
                #end k loop
                C[i][j] = sum
                j += 1
            #end j loop
            i += 1
        #end i loop
        return C
    #end if
    else:
        return False
    
def multiplybyfor(A,B):
    C = np.zeros(((len(A)),(len(B[0]))),dtype=int)
    if len(A) == len(B[0]):
        for i in range (0,len(A)):
            for j in range (0,len(B[0])):
                sum = 0
                for k in range (0,len(A)):
                    sum = sum + (A[i][k]*B[k][j])
                C[i][j] = sum
        return C
    else:
        return False
```

```python
def deter(x):
    det = 0
    if len(x) == 1:
        det = x[0][0]
    else:
        for i in range (0,len(x[0])):
            mi = minor(x,0,i)
            small = (mi[0][0]*mi[1][1])-(mi[0][1]*mi[1][0])
            det = det + ((-1)**i)*x[0][i]*small
    return det
```

```python
def fwd_elim(Ab):
    Ab_re = Ab.copy()
    [nr,nc] = Ab_re.shape
    
    # set i as the Iteration index
    for i in range(1,nr):
        print('Iteration #{}: '.format(i))
        # set r as the row index to work on
        for r in range(i,nr):
            print('  working on row index {}'.format(r))
            # iteration index represent the order of the variable that we want to eliminate
            row_factor = Ab_re[r,i-1]/Ab_re[i-1,i-1]
            Ab_re[r,:] = Ab_re[r,:] - Ab_re[i-1,:]*row_factor
        #end for r
    #end for i
    print(Ab_re)
    return Ab_re
#end function
```

```python
def back_sub(Ab_re):
    [nr,nc] = Ab_re.shape
    # nr will represent the total number of variables that we have to find
    # create x to hold calculated values
    x = np.empty(nr)
    
    # v is the variable iterator (index)
    for v in range(nr-1,-1,-1):
        print('v = {}'.format(v))
        # x_v = (constant - (sum_of_products between coeff_i and x_i))/coeff of x_v
        # where i > v
        
        sum_of_products = 0
        for i in range(v+1,nr):
            print('  i = {}'.format(i))
            product = Ab_re[v,i]*x[i]
            sum_of_products = sum_of_products + product
        #end for i
        
        x[v] = (Ab_re[v,nc-1] - sum_of_products)/Ab_re[v,v]
    #end for v
    print(x)
    return x

#end function
```
