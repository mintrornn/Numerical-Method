```python
import numpy as np
import math
```

```python
#Task1
def fmax(x):
    return x*(1-x)*np.e**x

def interval_search_max(fmax,a,b,p):
    es = 0.5*(10**(2-p))
    ea = 100
    prev_xmax = b
    
    i = 1

    while (ea > es):
        epsilon = (a+b)/10
        xmid = (a+b)/2
        alpha = xmid - (epsilon/2)
        beta = xmid + (epsilon/2)
        if (fmax(alpha) < fmax(beta)):
            a = alpha
            xmax = (a+b)/2
        else:
            b = beta
            xmax = (a+b)/2
        ea = abs((xmax - prev_xmax)/xmax) * 100  
        print('Iteration #{}: interval = [{},{}], xmax = {}, ea = {}'.format(i,a,b,xmax,ea)) 
        prev_xmax = xmax
        i += 1
        
interval_search_max(fmax,0,2,4)
```

```python
#Task2
def fmin(x):
    return (x**5)-3*x**4+5

def interval_search_min(fmin,a,b,p):
    es = 0.5*(10**(2-p))
    ea = 100
    prev_xmin = b
    
    i = 1    
    while (ea > es):
        epsilon = (a+b)/10
        xmid = (a+b)/2
        alpha = xmid + (epsilon/2)
        beta = xmid - (epsilon/2)
        if (fmin(beta)<fmin(alpha)):
            b = alpha
            xmin = (a+b)/2
        else:
            a = beta
            xmin = (a+b)/2
        ea = abs((xmin - prev_xmin)/xmin) * 100
        print('Iteration #{}: interval = [{},{}], xmin = {}, ea = {}'.format(i,a,b,xmin,ea))
        prev_xmin = xmin
        i += 1
        
interval_search_min(fmin,0,4,4)
```

```python
#Task3
def f(x):
    return x**5-3*x**4+5

def golden_min(f,xl,xu,p):
    es = 0.5*(10**(2-p))
    ea = 100
    
    i = 1
    phi = (1+math.sqrt(5))/2
    while (ea > es):
        d = (phi - 1)*(xu - xl)
        x1 = xl + d
        x2 = xu - d
        if (f(x1)<f(x2)):
            xl = x2
            xmin = x1
        else:
            xu = x1
            xmin = x2        
        ea = ((2 - phi)*(xu - xl)/xmin)*100
        print('Iteration #{}: interval = [{},{}], xmin = {}, ea = {}'.format(i,xl,xu,xmin,ea))
        i += 1

golden_min(fmin,0,4,4)
```

```python
#Task5

def f(x):
    return x**5-3*x**4+5

def parabolic_min(f,x1,x2,x3,p):
    es = 0.5 * (10**(2-p))
    ea = 100
    prev_xmax = x2
    i = 1
    
    while (ea > es):
        alpha1 = (x2-x1)*(x2-x1)*(f(x2)-f(x3))
        alpha2 = (x2-x3)*(x2-x3)*(f(x2)-f(x1))
        beta1 = (x2-x1)*(f(x2)-f(x3))
        beta2 = (x2-x3)*(f(x2)-f(x1))
        gamma = (alpha1 - alpha2)/(beta1 - beta2)

        x4 = x2 - (0.5 * gamma)
        
        if (x4 > x2):
            x3 = x2
            x2 = x4
            xmax = x4
        else:
            x1 = x2
            x2 = x4
            xmax = x4
        print('Iteration #{}: x4 = {}'.format(i,x4))
        ea = abs((xmax - prev_xmax)/xmax)*100
        print('Iteration #{}: [x1,x2,x3] = [{},{},{}], xmax = {}, ea = {}'.format(i,x1,x2,x3,xmax,ea))
        prev_xmax = xmax
        i += 1
        
parabolic_min(fmin,2.3, 2.3+2.5/1000, 2.5,4)
```
